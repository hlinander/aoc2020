```haskell
acc (d1, d3, p) n = 
  case (n - p) of
    1 -> (d1 + 1, d3, n)
    2 -> (d1, d3, n)
    3 -> (d1, d3 + 1, n)
```

```haskell
-- The naive, unmemoized version
combs l p = 
  let
    v = filter (\x -> x <= p + 3 && x > p) l
  in case v of
    [] -> 1
    y -> sum (map (combs l) y)
```

```haskell
-- Open recursion on cp
combsFast cp p = 
  let
    v = filter (\x -> x <= p + 3 && x > p) list
  in case v of
    [] -> 1
    y -> sum (map cp y)
```

```haskell
-- Memoized but linear time in indexing (https://stackoverflow.com/questions/3208258/memoization-in-haskell)
combsList = map (combsFast memComb) [1..]

memComb p = combsList !! p
```

```haskell
-- The cool tree version with log(n) indexing time

data Tree a = Tree (Tree a) a (Tree a)
instance Functor Tree where
  fmap f (Tree l m r) = Tree (fmap f l) (f m) (fmap f r)
  
index (Tree _ m _) 0 = m
index (Tree l _ r) n = case (n - 1) `divMod` 2 of
  (q, 0) -> index l q
  (q, 1) -> index r q
  
nats = go 0 1
  where
    go !n !s = Tree (go l s') n (go r s')
      where
        l = n + s
        r = l + s
        s' = s * 2
```

```haskell
combTree = fmap (combsFast memCombsTree) nats
memCombsTree = index (combTree)
```
